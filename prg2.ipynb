```python
# Design a single unit perceptron for classification of linearly separable binary dataset without using pre-defined models. Use the perceptron from sklearn 
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.linear_model import Perceptron

# Generate dataset
X, y = make_blobs(n_samples=100, centers=2, random_state=42)

# Train Perceptron
model = Perceptron(max_iter=1000, eta0=0.01, random_state=42).fit(X, y)

# Plot data and decision boundary
def plot_decision_boundary(model, X, y):
    xx, yy = np.meshgrid(
        np.linspace(X[:, 0].min() - 1, X[:, 0].max() + 1, 200),
        np.linspace(X[:, 1].min() - 1, X[:, 1].max() + 1, 200)
    )
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolors='k')
    plt.title('Perceptron Decision Boundary')
    plt.show()

plot_decision_boundary(model, X, y)
```

```python
from sklearn.linear_model import Perceptron
import numpy as np
import matplotlib.pyplot as plt

X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_and = np.array([0, 0, 0, 1])
y_or = np.array([0, 1, 1, 1])
y_xor = np.array([0, 1, 1, 0])

# Create a perceptron model
def train_and_evaluate(X, y):
    perceptron = Perceptron(max_iter=1000, eta0=1, random_state=0)
    perceptron.fit(X, y)
    
    predictions = perceptron.predict(X)
    
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm')
    plt.show()

    accuracy = np.mean(predictions == y)
    print(f'Accuracy: {accuracy * 100:.2f}%')

print("Evaluating AND Dataset:")
train_and_evaluate(X, y_and)

print("\nEvaluating OR Dataset:")
train_and_evaluate(X, y_or)

print("\nEvaluating XOR Dataset:")
train_and_evaluate(X, y_xor)
```
